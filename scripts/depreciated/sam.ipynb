```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from segment_anything import SamAutomaticMaskGenerator, sam_model_registry
DATAFILE = "/scratchdata/far"
DEVICE = "cuda:0"
sam = sam_model_registry["vit_h"](checkpoint="/scratchdata/sam_vit_h_4b8939.pth")
#sam = sam_model_registry["vit_l"](checkpoint="/scratchdata/sam_vit_l_0b3195.pth")
sam.to(DEVICE)

mask_generator = SamAutomaticMaskGenerator(sam)
```

```python
img = cv2.imread(DATAFILE + "/color.png")
masks = mask_generator.generate(img)
for i in range(len(masks)):
    masks[i]["segmentation"] = masks[i]["segmentation"].astype(np.uint8).tolist()
```

```python
import json

with open(DATAFILE + "/masks.json", "w") as f:
    json.dump({"sam":masks}, f, indent=2)
```

```python
#Visualised segmented regions

seg_img = np.zeros((img.shape[0], img.shape[1]), dtype=np.uint8)

for i, mask in enumerate(masks):
    mask_region = mask["segmentation"]
    mask_region = np.array(mask_region, dtype=np.uint8)
    seg_img[mask_region == 1] = i

plt.imshow(seg_img)
```
