```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

DATAFILE = "/scratchdata/stationary"

#Read csv file
CSV = np.genfromtxt(os.path.join(DATAFILE, "accuracy.csv"), delimiter=',', skip_header=1)
```

```python
print(CSV)
```

```python
#Average time
print("Mean Time:", CSV[:,0].mean())
```

```python
#Average accuracy
#Set Figure size
fig, ax = plt.subplots(1,2)
fig.set_size_inches(15, 5)

ax[0].plot(CSV[:,1])
ax[0].plot(CSV[:,4])

ax[0].legend(["Raw", "Depth Anything"])
ax[0].set_title("Mean Error (m)")

ax[1].plot(CSV[:,2])
ax[1].plot(CSV[:,5])

ax[1].legend(["Raw", "Depth Anything"])
ax[1].set_title("Max Error (m)")
```
